## Find max Sharpe ratio portfolio

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import risk_kit as kit
```

```python
ind = kit.get_ind_returns()
er = kit.annualize_rets(ind["1996":"2000"], 12)
cov = ind["1996":"2000"].cov()
```

```python
kit.plot_ef(20, er, cov)
```

```python
ax = kit.plot_ef(20, er, cov)
ax.set_xlim(left=0)
```

```python
ax = kit.plot_ef(20, er, cov)
ax.set_xlim(left=0)
rf = 0.1
weights_MSR = kit.max_Sharpe_ratio(rf, er, cov)
returns_MSR = kit.portfolio_return(weights_MSR, er)
volatility_MSR = kit.portfolio_vol(weights_MSR, cov)
# Add CML
cml_x = [0, volatility_MSR]
cml_y = [rf, returns_MSR]
ax.plot(cml_x, cml_y, color="green", marker="o", linestyle="dashed")
```
